```python
#IMPORTING LIBRARIES
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,mean_squared_error
```

```python
data = pd.read_csv(r"C:\Users\Muruganandan\OneDrive\Desktop\MY PROJECTS\LSTM REGRESSION\GOOG1.csv")
df = pd.DataFrame(data)
df.info()
```

```python
#converting  date column to date time format 
df['date']=pd.to_datetime(df['date'])

```

```python
#creating new columns by shifting 1 row of target,2 rows of targets , mean of 3 rows of target and mean of 7 rows of target
df['close_lag1'] = df['close'].shift(1)
df['close_lag2'] = df['close'].shift(2)
df['rolling_3'] = df['close'].rolling(window=3).mean().shift(1)
df['rolling_7'] = df['close'].rolling(window=7).mean().shift(1)
```

```python
#seperating target and features
df['target'] = df['close']
features = ['close_lag1','close_lag2','rolling_3','rolling_7']

x = df[features]
y=df['target'].shift(-1)
dates =df['date']


x = x.fillna(x.mean())
y = y.fillna(y.mean())
#splitting training and testing data

train_size = int(len(df)*0.8)

x_train,x_test = x.iloc[:train_size],x.iloc[train_size:]
y_train,y_test = y.iloc[:train_size],y.iloc[train_size:]
test_dates = dates.iloc[train_size:]

# CHOOSING AND TRAINING THE MODEL

model = LinearRegression()
model.fit(x_train,y_train)
```

```python
#PREDICTING THE MODEL

y_pred = model.predict(x_test)

# evaluating the model

r2 = r2_score(y_test,y_pred)
mse = mean_squared_error(y_test,y_pred)

print("r2 is",r2)
print("mse is",mse)
```

```python
#obtaining exact value of mse
a=y.mean()
b= np.sqrt(mse)

correct_mse = b/a
correct_mse
```

```python
#visualisation of stock price predcitions

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12,6))
plt.plot(test_dates,y_test,label= 'actual')
plt.plot(test_dates,y_pred,label= 'actual')
plt.legend()
plt.title('PREDCITION VS ACTUAL')
plt.show()
```
